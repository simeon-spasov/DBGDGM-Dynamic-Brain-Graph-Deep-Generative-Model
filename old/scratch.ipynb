```python
import pickle as pkl
import matplotlib.pyplot as plt
import numpy as np
```

```python
roi_labels = np.loadtxt("./roi_labels.txt", dtype=str)[:-1]
roi_labels = [l.split("_ROI")[0] for l in roi_labels]

with open("static_comms_fmri.pkl", "rb") as file:
   results = pkl.load(file)

subject = results[0][0]
```

```python
num_community, num_roi = subject.shape
fig, axes = plt.subplots(num_community, 1, figsize=(20, 30), gridspec_kw = {"wspace": 0, "hspace": 0.30})
for idx, ax in enumerate(axes):
    ax.bar(x=range(num_roi), height=subject[idx, :])
    ax.set_ylabel("Probability")
    ax.set_xlabel("ROI")
    ax.set_xticks(range(num_roi))
    ax.set_xticklabels(roi_labels, rotation=90)
    ax.set_title("Community {}".format(idx))
plt.show()
```
